```python
import pandas as pd
import numpy as np

import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv('train_2v.csv')

```

```python
train.head()
```

```python
train['bmi'].fillna((train['bmi'].mean()), inplace=True)



```

```python
train["smoking_status"].fillna("missing", inplace = True) 
```

```python
train.head()
```

```python
train = pd.get_dummies(train)
```

```python
train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python


dtree = DecisionTreeClassifier()
```

```python
X_train = train.drop(['stroke'],axis=1)
```

```python
y_train = train['stroke']
```

```python
test = pd.read_csv('test_2v.csv') 
```

```python
 test['bmi'].fillna((test['bmi'].mean()), inplace=True)



```

```python
test["smoking_status"].fillna("missing", inplace = True) 
```

```python
test = pd.get_dummies(test)
```

```python
test.head()
```

```python
X_test = test
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
predictions
```

```python
X_test['strokes'] = predictions
```

```python
ultimate = pd.DataFrame()
```

```python
ultimate['id'] = X_test['id']
```

```python
ultimate['stroke'] = np.vectorize(lambda s: 'YES' if s==1 else 'NO')(predictions)
```

```python
ultimate.to_csv('ultimate_report.csv')
```
